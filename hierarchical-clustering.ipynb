```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    '身長': [160, 170, 175, 180],
    '体重': [60, 65, 80, 75]
})
```

```python
# データから距離行列を計算してlinkageを実行
Z = linkage(data, method='ward')

# デンドログラムの表示
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=['人1', '人2', '人3', '人4'])
plt.title('デンドログラム')
plt.xlabel('個人')
plt.ylabel('クラスタ間距離')
plt.show()
```
